```python
# Import Library yang diperlukan 
import matplotlib.pyplot as plt

from skimage import data
from skimage.io import imread
from skimage.color import rgb2gray 
import numpy as np
```

```python
# Load & Plot Input Image

# Membaca Citra
citra1 = imread(fname="Image/gedung.tif")  
citra2 = imread(fname="Image/goldhill.tif")  

# Menampilkan dimensi citra  dengan menggunakan atribut shape.
print('Shape citra 1 : ', citra1.shape)  
print('Shape citra 2 : ', citra2.shape)  

# Membuat subplot dengan 1 baris dan 2 kolom, serta mengatur ukuran figur menjadi 10x10.
fig, axes = plt.subplots(1, 2, figsize=(10, 10)) 
# Membuat array 1 dimensi dari axes. 
ax = axes.ravel()  

# Menampilkan citra  dalam axes menggunakan cmap gray dan memberikan title
ax[0].imshow(citra1, cmap='gray')  
ax[0].set_title("Citra 1")  
ax[1].imshow(citra2, cmap='gray')  
ax[1].set_title("Citra 2")  
```

```python
# Menyiapkan variable output

# Membuat salinan citra1 dan citra2 dan mengonversinya ke tipe data float.
copyCitra1 = citra1.copy().astype(float) 
copyCitra2 = citra2.copy().astype(float)  

# Mendapatkan dimensi citra1 
m1,n1 = copyCitra1.shape 
# Membuat array kosong untuk menyimpan output
output1 = np.empty([m1, n1])  

m2,n2 = copyCitra2.shape  # Mendapatkan dimensi citra2 
output2 = np.empty([m2, n2])  # Membuat array kosong menyimpan output.

print('Shape copy citra 1 : ', copyCitra1.shape)  # Menampilkan dimensi copyCitra1 
print('Shape output citra 1 : ', output1.shape)  # Menampilkan dimensi 
# Menampilkan nilai m1 dan n1
print('m1 : ',m1)  
print('n1 : ',n1)  
print()

print('Shape copy citra 2 : ', copyCitra2.shape)  # Menampilkan dimensi copyCitra2 
print('Shape output citra 3 : ', output2.shape)  # Menampilkan dimensi output2 
# Menampilkan nilai m2 dan n2
print('m2 : ',m2)  
print('n2 : ',n2)  
print()
```

```python
# Proses Filter Rerata Pada Citra Input 1

for baris in range(0, m1-1):  # Melakukan perulangan untuk setiap baris dari 0 hingga m1-2.
    for kolom in range(0, n1-1):  # Melakukan perulangan untuk setiap kolom dari 0 hingga n1-2.
        a1 = baris  # Menyimpan nilai baris dalam variabel a1.
        b1 = kolom  # Menyimpan nilai kolom dalam variabel b1.
        jumlah = copyCitra1[a1-1, b1-1] + copyCitra1[a1-1, b1] + copyCitra1[a1-1, b1-1] + \
                 copyCitra1[a1, b1-1] + copyCitra1[a1, b1] + copyCitra1[a1, b1+1] + \
                 copyCitra1[a1+1, b1-1] + copyCitra1[a1+1, b1] + copyCitra1[a1+1, b1+1];  # Menghitung jumlah dari 9 tetangga piksel.
        output1[a1, b1] = (1/9 * jumlah)  # Menyimpan hasil rata-rata dari tetangga piksel dalam output1[a1, b1].
```

```python
# Proses Filter Rerata Pada Citra Input 2

# komennya sama kayak yg atas Citra input 1
for baris1 in range(0, m2-1):  
    for kolom1 in range(0, n2-1):  
        a1 = baris1  
        b1 = kolom1  
        jumlah = copyCitra2[a1-1, b1-1] + copyCitra2[a1-1, b1] + copyCitra2[a1-1, b1-1] + \
                 copyCitra2[a1, b1-1] + copyCitra2[a1, b1] + copyCitra2[a1, b1+1] + \
                 copyCitra2[a1+1, b1-1] + copyCitra2[a1+1, b1] + copyCitra2[a1+1, b1+1];  
        output2[a1, b1] = (1/9 * jumlah)  
```

```python
# Plot Citra Input dan Output Hasil dari Filter Rerata

# Membuat subplots dengan 2 baris dan 2 kolom, dengan ukuran gambar 10x10.
fig, axes = plt.subplots(2, 2, figsize=(8, 8))  
# Meratakan array axes menjadi satu dimensi.
ax = axes.ravel() 

# Menampilkan citra1 pada axis 0 samapi axis 3 dengan colormap 'gray' dan memberikan title
ax[0].imshow(citra1, cmap='gray')  
ax[0].set_title("Input Citra 1")  

ax[1].imshow(citra2, cmap='gray')  
ax[1].set_title("Input Citra 1")  

ax[2].imshow(output1, cmap='gray')  
ax[2].set_title("Output Citra 1")  

ax[3].imshow(output2, cmap='gray')  
ax[3].set_title("Output Citra 2")  
```
