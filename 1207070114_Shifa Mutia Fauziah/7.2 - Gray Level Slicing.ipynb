```python
# Import Library yang diperlukan 
import cv2
import numpy as np
from skimage import data
import matplotlib.pyplot as plt
```

```python
# Membaca Gambar
img = cv2.imread("Image/mulan.jpg")  
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) #untuk mengubah citra img dari format BGR menjadi grayscale. 
```

```python
# Mendapatkan dimensi baris dan kolom citra 
row, column = img.shape  

# Membuat array kosong
img1 = np.zeros((row,column),dtype = 'uint8')   

# Mengatur nilai batas bawah (minimum) dan nilai batas atas (maksimum) untuk rentang intensitas piksel.
min_range = 10  
max_range = 60  

for i in range(row):  # Melakukan iterasi untuk setiap baris dalam citra.
    for j in range(column):  # Melakukan iterasi untuk setiap kolom dalam citra.
        if img[i,j]>min_range and img[i,j]<max_range:  # Memeriksa apakah intensitas piksel
            img1[i,j] = 255  # Jika iya, mengatur intensitas piksel pada posisi (i, j) dalam img1 menjadi 255 (putih).
        else:
            img1[i,j] = 0  # Jika tidak, mengatur intensitas piksel pada posisi (i, j) dalam img1 menjadi 0 (hitam).
```

```python
# Plot Image

# Membuat subplot 
fig, axes = plt.subplots(2, 2, figsize=(12, 12))  

# Mengubah array 2D axes menjadi array 1D menggunakan ravel()
ax = axes.ravel()  

# Menampilkan citra input dalam axes pertama 
ax[0].imshow(img, cmap=plt.cm.gray)  
ax[0].set_title("Citra Input")  

# Menampilkan histogram citra input dalam axes kedua .
ax[1].hist(img.ravel(), bins=256)  
ax[1].set_title('Histogram Input')  

# Menampilkan citra output dalam axes ketiga 
ax[2].imshow(img1, cmap=plt.cm.gray)  
ax[2].set_title("Citra Output")  

# Menampilkan histogram citra output dalam axes keempat 
ax[3].hist(img1.ravel(), bins=256) 
ax[3].set_title('Histogram Output')  
```
